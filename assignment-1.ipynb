```python
# Meaningful visualizations using the dataset.
```

```python
plt.figure()
plt.hist(df['SL'])
plt.title('data of sepal length')
plt.xticks()
plt.yticks()
```

```python
plt.figure()
plt.hist(df['SW'])
plt.title('data of sepal width')
plt.xticks()
plt.yticks()
```

```python
plt.boxplot(df['PW'])
plt.title('box plot of petal width')
plt.xlabel('Petal width')
```

```python
sns.boxplot(x='PW',y='PL',data=df)
```

```python
sns.pairplot(df)
```

```python
plt.figure()
plt.scatter(df['SL'],df['SW'],s=5,c='red')
plt.title('plot of SL vs SW')
plt.xlabel('SL')
plt.ylabel('SW')
plt.xticks
plt.yticks
```
